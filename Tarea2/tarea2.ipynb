#### Let's start by importing all the necessary tools we need

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.stats import binom
```

```python
rows = 6
bins = rows + 1
R = 10000
```

```python
rsultados = np.random.binomial(rows, 0.5, size =R)
freq, _ = np.histogram(rsultados, bins = bins, range = (0, rows))
freq_rel = freq / R
```

```python
x = np.arange(0, bins)
prob = binom.pmf(x, rows, 0.5)
```

```python
plt.figure(figsize=(8,8))
plt.bar(x, freq_rel, width=0.6, alpha = 0.6, label = 'Simulation')
plt.plot(x , prob, 'ro-', label = 'Theoretical binomial distribution', markersize = 8)
plt.xlabel('final bins')
plt.ylabel('probability')
plt.xticks(x)
plt.legend()
plt.grid(True, linestyle = '--', alpha = 0.6)
plt.show()
```
